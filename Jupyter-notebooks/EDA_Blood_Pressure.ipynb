```python
import bokeh
from bokeh.io import output_notebook,show,curdoc 
from bokeh.models import Select,ColumnDataSource
from bokeh.palettes import Spectral10
from bokeh.plotting import figure
from bokeh.layouts import row,column,widgetbox
from bokeh.application.handlers import FunctionHandler
from bokeh.application import Application
output_notebook()
```

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\MAMTA\Downloads\Python Intro and Stats\Probability and Stats\blood_pressure.csv')
df.head()

# data cleanup
df['bp_before']=df['bp_before'].astype(str)
df['bp_after']=df['bp_after'].astype(str)


COLORS = Spectral10
df.head()
```

```python
def modify_doc(doc):
    
    def make_dataframe(age,gender):
        bp_Before = []
        bp_After = []
        Age=[]
        Gender = []
        
        new_df = df.copy()
        if(age!='All'):
            new_df = new_df[new_df['agegrp']==age]
        if(gender!='Both'):
            new_df = new_df[new_df['sex']==gender]

        for index,row in new_df.iterrows():
            bp_Before.append(row['bp_before'])
            bp_After.append(row['bp_after'])
            Age.append(row['agegrp'])
            Gender.append(row['sex'])
            

        new_df = pd.DataFrame(data = {'bp_before':bp_Before,'bp_after':bp_After,'agegrp':Age,'sex':Gender})
        return new_df

    def make_plot(cds):
        p = figure(plot_width = 600, plot_height = 400, title = 'BP before vs after',x_axis_label = 'BP Before',y_axis_label = 'BP After')
        
        colors = []
        
        if Color.value != 'None':        
            column_to_use = Color.value
            x = list(np.random.choice(COLORS,cds[Color.value].nunique()))
            column_entries = list(cds[Color.value].unique())

            for temp in cds[Color.value].tolist():
                colors.append(x[column_entries.index(temp)])
            cds['color'] = pd.Series(colors)
            p.circle(x='bp_before',y='bp_after',color = 'color',source=cds,size=10)
       
            return p
        else :
            p.circle(x='bp_before',y='bp_after',source=cds,size=10)
            return p

    def update(attrs,old,new):
        new_df = make_dataframe(Age.value,Gender.value)
        layout.children[1] = make_plot(new_df)

    
    
    Age = Select(title='Age Group',value = '30-45',options=list(df.agegrp.unique())+['All'])
    Age.on_change('value',update)

    Gender = Select(title='Sex',value='Male',options=list(df.sex.unique())+['Both'])
    Gender.on_change('value',update)
    
    Color = Select(title='Color',value='sex',options=['None','agegrp','sex'])
    Color.on_change('value',update)
    
    controls = widgetbox(row(Age,Gender,Color))
    x = make_dataframe(Age.value,Gender.value)
    p = make_plot(x)
    layout = column(controls,p)
    doc.add_root(layout)
```

```python
handler = FunctionHandler(modify_doc)
app = Application(handler)
```

```python
show(app)
```
